```python
import folium
from folium.plugins import HeatMap

m=folium.Map(location=[10.7565577, 106.6248006], zoom_start=15)
folium.Marker([10.7565577, 106.6248006], popup='Giao thông ùn tắc nhẹ', tooltip='My home').add_to(m)
folium.Marker([10.7536977, 106.6346337], popup='Siêu siêu siêu siêu tắc đường', tooltip='Bung Binh Phu Lam').add_to(m)
folium.Marker([10.7532397, 106.6335908], popup='Sieu Cấp Tắc Đường', tooltip='Vòng xoay Phú Lâm').add_to(m)
folium.Marker([10.7551474, 106.6342982], popup='Tắc đường vào giờ tan học', tooltip='Tiểu Học Phú Lâm').add_to(m)
heat_data=[[10.7551474, 106.6342982, 0.9], [10.7532397, 106.6335908, 0.5], [10.7565577, 106.6248006, 0.2], [10.7536977, 106.6346337, 1.0]]
HeatMap(heat_data).add_to(m)

m
```
