```python
import folium
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
import time


geolocator = Nominatim(user_agent="my_exercise_app")

dia_diem = [
    "972 An Duong Vuong, Ho Chi Minh City",
    "UEH University A, Ho Chi Minh City",
    "Ho Con Rua, Ho Chi Minh City",
    "Regina Mundi Church, Ho Chi Minh City",
    "San Khau Kich 5B, Ho Chi Minh City",
    "So Ngoai Vu TP.HCM",
    "THPT Marie Curie Ho Chi Minh",
    "Cong vien Dam Sen",
    "Vong xoay Phu Lam",
    "220 Au Co Ho Chi Minh",
    "San bong Phu Tho",
]

ds_toa_do = []

for i in dia_diem:
    location = geolocator.geocode(i)
    if location:
        ds_toa_do.append((i, (location.latitude, location.longitude)))
    else:
        print(f"Không tìm thấy tọa độ cho: {i}")
    time.sleep(1)

if ds_toa_do:
    nha_coords = ds_toa_do[0][1]
    nha_ten = ds_toa_do[0][0]

    m = folium.Map(location=nha_coords, zoom_start=14)
    folium.Marker(
        location=nha_coords,
        popup="Nhà của Hoàng Đình Nguyên dep zai vip pro",
        icon=folium.Icon(color="red", icon="home"),
        tooltip='My Home'
    ).add_to(m)

    for name, coord in ds_toa_do[1:]:
        khoang_cach = geodesic(nha_coords, coord).kilometers
        khoangcach = f"{name}: cách nhà {khoang_cach:.2f} km"

        folium.Marker(
            location=coord,
            popup=khoangcach,
            tooltip=name,
        ).add_to(m)

        folium.PolyLine(locations=[nha_coords, coord], color='blue', weight=2, opacity=0.7).add_to(m)

    display(m)
```
